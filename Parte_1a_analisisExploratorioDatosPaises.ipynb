```python
'''
Rank:Indica la posición del país o territorio en un ranking global de densidad poblacional. Los valores son números enteros. 
Los guiones ("–") indican que no hay una clasificación oficial, pero se incluye el territorio para contexto.
Interpretación: Este ranking muestra la comparación relativa entre países o territorios en términos de densidad poblacional. 
Los países con menores superficies y alta población tienden a ocupar los primeros lugares (como Macau y Monaco).

Country (or dependent territory): Nombre del país o territorio dependiente al que se refiere el registro.
Interpretación: Cada fila corresponde a un país o territorio específico que aparece en el ranking según su densidad de población. Algunos territorios, como Hong Kong o Bermuda, son dependientes de otros países, y este campo permite identificarlos claramente.

Area km²: Área total del país o territorio expresada en kilómetros cuadrados (km²).
Interpretación: Este dato indica la extensión física del país o territorio. Las áreas más pequeñas tienden a tener densidades poblacionales más altas si la población es significativa, como en Macau.

Area mi²: Área total expresada en millas cuadradas (mi²), la unidad de medida utilizada en países anglosajones.
Interpretación: Esta columna es útil para audiencias que prefieren las unidades imperiales para interpretar áreas geográficas.

Population: Población total del país o territorio en la fecha indicada.
Interpretación: Este dato muestra el número total de personas que viven en el país o territorio en el momento del informe.

Density pop./km²: Densidad de población por kilómetro cuadrado. Se calcula dividiendo la población total entre el área en km².
Interpretación: Una densidad alta sugiere un territorio muy poblado en relación con su superficie, lo que puede generar congestión o demanda alta de servicios públicos.

Density pop./mi²: Densidad de población por milla cuadrada, una unidad alternativa.
Interpretación: Al igual que la densidad por km², esta métrica refleja la población por unidad de área, pero en millas cuadradas.

Date: Fecha del informe o estimación poblacional utilizada.
Interpretación: Esta columna permite determinar la actualidad del dato y si los valores reflejan una proyección o una medición real.
Ejemplos:

Population source: Indica la fuente de los datos de población, como informes oficiales o proyecciones.
Interpretación: La fuente puede ser un censo oficial, una estimación anual, o una proyección de la ONU. Esto es importante para evaluar la fiabilidad del dato poblacional.


Relación entre las columnas: Las columnas Area km², Population, y Density se relacionan directamente:
Density = Population/Area
 
Fechas recientes indican datos más fiables y actuales.
Territorios pequeños con alta densidad poblacional (como Macau y Monaco) muestran desafíos específicos, como limitaciones de espacio y mayor demanda de servicios.

'''
```

```python
# Eliminar filas con nulos en la columna 'Population source'
df_Pupulation_density_by_countries_sinNan = df_Pupulation_density_by_countries.dropna(subset=['Population source']).copy()

print(f"Filas después de eliminar nulos en 'Area km2': {len(df_Pupulation_density_by_countries_sinNan)}")
```

```python
##Ya que date tiene un formato atípico, quiero cambiar el formato y agregar columnas Day, Month y Year
# Convierto las fechas a formato 'DD-MM-YYYY'
df_Pupulation_density_by_countries_sinNan_fechas = df_Pupulation_density_by_countries_sinNan.copy()

df_Pupulation_density_by_countries_sinNan_fechas['Date'] = pd.to_datetime(
    df_Pupulation_density_by_countries_sinNan_fechas['Date'], 
    format='%B %d, %Y',  # Cambiar al formato correcto
    errors='raise'  # Detenerse si alguna conversión falla
)

# Creo columnas 'Day', 'Month' y 'Year'
df_Pupulation_density_by_countries_sinNan_fechas['Day'] = df_Pupulation_density_by_countries_sinNan_fechas['Date'].dt.day
df_Pupulation_density_by_countries_sinNan_fechas['Month'] = df_Pupulation_density_by_countries_sinNan_fechas['Date'].dt.month
df_Pupulation_density_by_countries_sinNan_fechas['Year'] = df_Pupulation_density_by_countries_sinNan_fechas['Date'].dt.year

# Muestro resultado
df_Pupulation_density_by_countries_sinNan_fechas
```

```python
# Copiar el DataFrame original
df_Pupulation_density_by_countries_sinNan_fechas_float = df_Pupulation_density_by_countries_sinNan_fechas.copy()

# Función para limpiar números grandes con comas cada tres dígitos (población, densidad)
def clean_large_numbers(value):
    if pd.isna(value):
        return np.nan
    try:
        # Eliminar todas las comas y convertir a float
        cleaned_value = float(value.replace(',', ''))
        return cleaned_value
    except ValueError:
        print(f"Valor no convertible a float: {value}")
        return np.nan

# Aplicar la limpieza en las columnas relevantes
for col in ['Population', 'Density pop./km2', 'Density pop./mi2']:
   df_Pupulation_density_by_countries_sinNan_fechas_float[col] = (
        df_Pupulation_density_by_countries_sinNan_fechas_float[col].apply(clean_large_numbers)
    )

def clean_area_values(value):
    if pd.isna(value):
        return np.nan
    try:
        # Eliminar TODAS las comas y convertir directamente a float
        cleaned_value = float(value.replace(',', ''))
        return cleaned_value
    except ValueError:
        print(f"Valor no convertible a float: {value}")
        return np.nan

# Aplicar la limpieza en las columnas de área
for col in ['Area km2', 'Area mi2']:
    df_Pupulation_density_by_countries_sinNan_fechas_float[col] = (
        df_Pupulation_density_by_countries_sinNan_fechas_float[col].apply(clean_area_values)
    )

# Mostrar el DataFrame final
df_Pupulation_density_by_countries_sinNan_fechas_float.head(10)
```
